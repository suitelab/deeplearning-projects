```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def function(x):
    a = 0.5
    b = 0.1
    
    return a * x + b
```

```python
def grad_function(x, h):
    f1 = function(x + h)
    f = function(x)
    return (f1 - f) / h
```

```python
def show_graph(x, f):
    y = f(x)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.plot(x, y)
    plt.show()
```

```python
x = np.arange(0.0, 1.0, 0.1)
show_graph(x, function)
```

```python
function(1)
```

```python
function(1.1)
```

```python
def sigmoid_function(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
def get_grad_sigmoid(x):
    f = sigmoid_function(x)
    return f * (1.0 - f)
```

```python
get_grad_sigmoid(1)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
show_graph(x, get_grad_sigmoid)
```
